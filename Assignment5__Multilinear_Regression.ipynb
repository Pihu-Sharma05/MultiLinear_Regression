# **50_Startups Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api  as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
from google.colab import files
uploaded= files.upload()
```

```python
df=pd.read_csv("50_Startups.csv")
df.head()
```

```python
df.info()
```

```python
df1=df.rename({'R&D Spend':'RDS','Administration':'ADMS','Marketing Spend':'MKTS'},axis=1)
df1.head()
```

```python
df1.describe()
```

```python
# correlation 
df1.corr()
```

```python
# Model Building
model=smf.ols("Profit~RDS+ADMS+MKTS",data=df1).fit()
```

```python
model.params
```

```python
model.tvalues, model.pvalues
```

```python
model.rsquared, model.rsquared_adj
```

```python
model_adms=smf.ols("Profit~ADMS",data=df1).fit()
model_adms.tvalues,model_adms.pvalues
```

```python
model_mkts=smf.ols("Profit~MKTS",data=df1).fit()
model_mkts.tvalues,model_mkts.pvalues
```

```python
model_admk=smf.ols("Profit~ADMS+MKTS",data=df1).fit()
model_admk.tvalues,model_admk.pvalues
```

```python
#Model Validation: Collinearity and Residual Analysis
rsqr_a=smf.ols("ADMS~RDS+MKTS",data=df1).fit().rsquared
vif_a=1/(1-rsqr_a)

rsqr_m=smf.ols("MKTS~RDS+ADMS",data=df1).fit().rsquared
vif_m=1/(1-rsqr_m)


rsqr_r=smf.ols("RDS~MKTS+ADMS",data=df1).fit().rsquared
vif_r=1/(1-rsqr_r)

d1={'Variables':['ADMS','MKTS','RDS'],'VIF':[vif_a,vif_m,vif_r]}
Vif_df=pd.DataFrame(d1)

Vif_df
```

There is no Collinearity because all variable has VIF<20

```python
#Residual Analysis
sm.qqplot(model.resid, line='q')
plt.title("Normal Q-Q Plot of Residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
# Test for Homoscedasticity or Heteroscedasticity (plotting model's standardized fitted values vs standardized residual values)

def standard_values(vals) : return (vals-vals.mean())/vals.std()  # User defined z = (x - mu)/sigma
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'RDS',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'ADMS',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'MKTS',fig=fig)
plt.show()
```

```python
#Model Deletion Diagnostics 
# Get influencers using cook's distance
(c,_)=model.get_influence().cooks_distance
c
```

```python
# Plot the influencers using the stem plot
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(df1)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

```python
k=df1.shape[1]
n=df1.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
# Remove the data points which are influencers and reassign the row number (reset_index(drop=True))
df2=df1.drop(df1.index[[49]],axis=0).reset_index(drop=True)
df2
```

```python
while np.max(c)>0.5 :
    model=smf.ols("Profit~RDS+ADMS+MKTS",data=df2).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    df2=df2.drop(df2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    df2
else:
    final_model=smf.ols("Profit~RDS+ADMS+MKTS",data=df2).fit()
    final_model.rsquared , final_model.aic
    print("Model accuracy",final_model.rsquared)
```

```python
#Model pridiction
new_data=pd.DataFrame({'RDS':70000,"ADMS":90000,"MKTS":140000},index=[0])
new_data
```

```python
final_model.predict(new_data)
```

```python
d2={'Prep_Models':['Model','Final_Model'],'Rsquared':[model.rsquared,final_model.rsquared]}
table=pd.DataFrame(d2)
table
```
